```
import pandas as pd
data=pd.read_csv("car-sales.csv")
data
```

```
data["Price"].plot()
```

```
data["Price"]=data["Price"].str.replace("[\$\,\.]","")
data["Price"]=data["Price"].str[:-2]
data["Total Sales"]=data["Price"].astype(int).cumsum()
data["Sale Data"]=pd.date_range("13/1/2020",periods=len(data))
data
```

```
data.plot(x="Sale Data", y="Total Sales")
```

```
data["Price"]=data["Price"].astype(int)
data.plot(x="Odometer (KM)", y="Price", kind="scatter")
```

```
heart=pd.read_csv("heart-disease.csv")
heart.head(10)
```

```
heart["age"].plot.hist(bins=50);
```

```
heart.plot.hist(subplots=True);
```

```
heart.plot.hist(figsize=(10,30), subplots=True);
```

```
over_50=heart[heart["age"]>50]
fig, ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(over_50["age"],
                   over_50["chol"],
                   c=over_50["target"])
ax.set(title="Heart Disease and Cholestrol Analysis",
       xlabel="Age",
       ylabel="Cholestrol");
ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```
fig, ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(over_50["age"],
                   over_50["chol"],
                   c=over_50["target"],
                   cmap="winter")
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol");
ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(over_50["chol"].mean(),
           linestyle="--",
           color="red");
```
